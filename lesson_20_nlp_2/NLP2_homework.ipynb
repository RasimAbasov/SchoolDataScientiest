# NLP2_4 (дополнительно) Реализовать классификатор токсичных комментариев tfidf на базе датасета (если не успели на классном занятии)

TF-IDF - это статистическая мера, которая оценивает важность слова в документе, который является частью коллекции документов. Эта мера учитывает частоту слова в документе и обратную частоту его встречаемости.

TfidfTransformer принимает на вход матрицу счетчиков слов, созданную с помощью CountVectorizer, и преобразует ее в матрицу векторов TF-IDF.

```python
import pandas as pd
from nltk.tokenize import WordPunctTokenizer
from nltk.stem import SnowballStemmer
from pymorphy2 import MorphAnalyzer


df = pd.read_csv('labeled.csv')

# создание экземпляра токенизатора
tokenizer = WordPunctTokenizer()

# создание экземпляра стеммера
stemmer = SnowballStemmer(language="russian")

# создание экземпляра лемматизатора
lemmatizer = MorphAnalyzer()


# Функция для stemming
def stem_tokens(tokens):
    stemmed = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed)

# Функция для lemmatization
def lemmatize_tokens(tokens):
    lemmatized = [lemmatizer.parse(token)[0].normal_form for token in tokens]
    return ' '.join(lemmatized)

# Применение токенизации и stemming к текстовому столбцу 'comment'
df['stemmed_text'] = df['comment'].apply(lambda x: stem_tokens(tokenizer.tokenize(x)))
df['lemmatized_text'] = df['comment'].apply(lambda x: lemmatize_tokens(tokenizer.tokenize(x)))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# создание BoW
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['lemmatized_text'])

# преобразование BoW в DataFrame
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
bow_df
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(bow_df)
df_idf = pd.DataFrame(tfidf_transformer.idf_, index=vectorizer.get_feature_names_out(), columns=["idf_weights"])
```
