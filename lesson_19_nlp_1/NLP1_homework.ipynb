# ML1_2: https://www.hackerrank.com/challenges/branch-reset-groups/problem?isFullScreen=true

Task

You have a test string .
Your task is to write a regex which will match , with following condition(s):

S consists of 8 digits.
S must have "---", "-", "." or ":" separator such that string S gets divided in 4 parts, with each part having exactly two digits.
S string must have exactly one kind of separator.
Separators must have integers on both sides.

Valid S

12-34-56-78
12:34:56:78
12---34---56---78
12.34.56.78


Invalid S

1-234-56-78
12-45.78:10

```python
import re
S = '12-34-56-78'
regex = r'^\d{2}([-.\:\s])\d{2}\1\d{2}\1\d{2}$'
reg = re.search(regex, S)
bool(reg)
```

# ML1_3: https://www.hackerrank.com/challenges/detect-html-links/problem?isFullScreen=true

Sample Input:1

2
<p><a href="http://www.quackit.com/html/tutorial/html_links.cfm">Example Link</a></p>
<div class="more-info"><a href="http://www.quackit.com/html/examples/html_links_examples.cfm">More Link Examples...</a></div>


Sample Output:1

http://www.quackit.com/html/tutorial/html_links.cfm,Example Link
http://www.quackit.com/html/examples/html_links_examples.cfm,More Link Examples...

```python
import re

html = '''2
<p><a href="http://www.quackit.com/html/tutorial/html_links.cfm">Example Link</a></p>
<div class="more-info"><a href="http://www.quackit.com/html/examples/html_links_examples.cfm">More Link Examples...</a></div>'''

regex = r'<a href="(.+?)">(.+?)</a>'
matches = re.findall(regex, html)

for match in matches:
    print(f'{match[0]},{match[1]}')
```

# ML1_4: Реализовать stemming, lemmatization & BoW на следующем датасете: https://cloud.mail.ru/public/Z4L3/vB8GcgTtK (Russian Toxic-abuse comments)

Stemming - это процесс нахождения основы слова путем удаления окончаний (например, суффиксов) без изменения основы слова. Он используется для уменьшения размерности данных и уменьшения числа уникальных слов в тексте.

Lemmatization - это процесс приведения слова к его базовой форме (лемме). Лемматизация учитывает морфологические характеристики слова, такие как падеж, число и время, чтобы получить его корневую форму. Например, слова "бежать", "бежит", "бежал" будут лемматизированы до слова "бежать".

BoW (Bag of Words) - это метод представления текста в виде вектора, где каждый элемент этого вектора соответствует количеству вхождений определенного слова (терма) в текст. В этом подходе учитывается только наличие или отсутствие термов в тексте, а не их порядок или контекст, что делает его простым и эффективным. Этот подход широко используется в задачах обработки естественного языка, таких как классификация текстов и построение тематических моделей.

```python
import pandas as pd
from nltk.tokenize import WordPunctTokenizer
from nltk.stem import SnowballStemmer
from pymorphy2 import MorphAnalyzer


df = pd.read_csv('labeled.csv')

# создание экземпляра токенизатора
tokenizer = WordPunctTokenizer()

# создание экземпляра стеммера
stemmer = SnowballStemmer(language="russian")

# создание экземпляра лемматизатора
lemmatizer = MorphAnalyzer()


# Функция для stemming
def stem_tokens(tokens):
    stemmed = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed)

# Функция для lemmatization
def lemmatize_tokens(tokens):
    lemmatized = [lemmatizer.parse(token)[0].normal_form for token in tokens]
    return ' '.join(lemmatized)

# Применение токенизации и stemming к текстовому столбцу 'comment'
df['stemmed_text'] = df['comment'].apply(lambda x: stem_tokens(tokenizer.tokenize(x)))
df['lemmatized_text'] = df['comment'].apply(lambda x: lemmatize_tokens(tokenizer.tokenize(x)))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# создание BoW
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['lemmatized_text'])

# преобразование BoW в DataFrame
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```
